```python
%sql
SELECT city, SUM(distinct_hotels) as total_hotels
FROM gold.hotel_weather_metrics
GROUP BY city
ORDER BY total_hotels DESC
LIMIT 5
```

```python
%sql
SELECT
  wthr_date,
  SUM(distinct_hotels) as number_of_reported_hotels,
  average_temperature_c AS avg_tmpr_c,
  max_temperature_c as max_tmpr_c,
  min_temperature_c as min_tmpr_c
FROM gold.hotel_weather_metrics
WHERE city = 'London'
GROUP BY wthr_date, average_temperature_c, max_temperature_c, min_temperature_c
```
