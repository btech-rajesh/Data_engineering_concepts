trimming whitespace from string fields

```python
import pyspark.sql.functions as F
import pyspark.sql.types as T

def trim_all_string_columns(df, trim_nbsp=True):
    """
    Trims whitespace from all string columns in a DataFrame.
    Optionally removes non-breaking spaces (\u00A0).
    """

    NBSP = "\u00A0"
    output_df = df

    for field in df.schema.fields:
        if isinstance(field.dataType, T.StringType):
            column_name = field.name
            col_expr = F.col(column_name)

            if trim_nbsp:
                col_expr = F.regexp_replace(col_expr, NBSP, " ")

            output_df = output_df.withColumn(
                column_name,
                F.trim(col_expr)
            )

    return output_df
transformed_df = trim_all_string_columns(dencrypted_df)

```

casting data from string to int using cast()

```python
from pyspark.sql.functions import col

def safe_cast(df, cast_map):
    """
    Casts columns according to cast_map only if:
      - The column exists in the DataFrame, and
      - The current type differs from the target type
    """
    dtypes = dict(df.dtypes)
    for c, target_t in cast_map.items():
        if c in dtypes:
            current_t = dtypes[c]
            if current_t != target_t:
                df = df.withColumn(c, col(c).cast(target_t))
                print(f"✔️ Casted {c}: {current_t} → {target_t}")
            else:
                print(f"ℹ️ {c} already {target_t}, skipping.")
        else:
            print(f"⚠️ Column {c} not found, skipping.")
    return df

# Define the desired target types for your schema
cast_plan = {
    # numeric doubles
    "avg_tmpr_c": "double",
    "avg_tmpr_f": "double",
    "latitude": "double",
    "longitude": "double",

    # integers
    "year": "int",
    "month": "int",
    "day": "int",

    # strings (usually already string, but included for completeness)
    "address": "string",
    "city": "string",
    "country": "string",
    "geoHash": "string",
    "id": "long",
    "name": "string",
    "wthr_date": "date",
    "_rescued_data": "string",
}

# Apply casts
transformed_df = safe_cast(transformed_df, cast_plan)

```

```python
def swap_name_address(df):
    

    col1 = "name"
    col2 = "address"
    temp_col = "__temp_col__"

    # Safety check
    if col1 not in df.columns or col2 not in df.columns:
        raise ValueError("Columns 'name' and/or 'address' not found in DataFrame")

    df = df.withColumnRenamed(col1, temp_col)
    df = df.withColumnRenamed(col2, col1)
    df = df.withColumnRenamed(temp_col, col2)

    return df
transformed_df=swap_name_address(transformed_df)
```

Removing records with null values in critical fields (e.g., using filter() or dropna()).

```python

def drop_if_null_or_blank(df, critical_cols, treat_blank_as_null=True, trim_nbsp=True):
    """
    Remove rows where any critical column is null/blank.
    - treat_blank_as_null: considers '' or whitespace-only as null equivalents for string cols.
    - trim_nbsp: also normalizes non-breaking space \u00A0 before trimming.
    """
    predicates = []
    NBSP = "\u00A0"

    for c in critical_cols:
        if c not in df.columns:
            raise KeyError(f"Critical column '{c}' not present in DataFrame")

        dtype = df.schema[c].dataType
        base_not_null = F.col(c).isNotNull()

        if treat_blank_as_null and isinstance(dtype, T.StringType):
            x = F.col(c)
            if trim_nbsp:
                x = F.regexp_replace(x, NBSP, " ")
            not_blank = F.length(F.trim(x)) > 0
            predicates.append(base_not_null & not_blank)
        else:
            predicates.append(base_not_null)

    # Keep only rows satisfying all predicates
    condition = F.lit(True)
    for p in predicates:
        condition = condition & p

    return df.filter(condition)
col1=["name","address"]
transformed_df=drop_if_null_or_blank(transformed_df, col1)
```

Removing duplicate records using deduplication logic compatible with streaming (e.g., dropDuplicates(["id", "timestamp_column"])).

```python

def normalize_blank_strings(df, cols):
    """
    Optional: standardize blanks to nulls for string columns in 'cols'.
    """
    NBSP = "\u00A0"
    out = df
    for c in cols:
        if c in df.columns and isinstance(df.schema[c].dataType, T.StringType):
            normalized = F.regexp_replace(F.col(c), NBSP, " ")
            out = out.withColumn(
                c,
                F.when(F.length(F.trim(normalized)) == 0, F.lit(None).cast(T.StringType()))
                 .otherwise(F.col(c))
            )
    return out

transformed_df=normalize_blank_strings(transformed_df,["id"])
```
