```python
# Storage and paths
STORAGE_ACCOUNT = "stdevwesteuropepqa5"
CONTAINER_NAME = "data"
BASE_PATH = f"wasbs://{CONTAINER_NAME}@{STORAGE_ACCOUNT}.blob.core.windows.net"
HOTEL_WEATHER_SOURCE_PATH = f"{BASE_PATH}/hotel-weather"

# Auto Loader metadata locations (in DBFS)
SCHEMA_LOCATION = f"{BASE_PATH}/schema/gold"
CHECKPOINT_LOCATION = f"{BASE_PATH}/checkpoint/gold"

# Read hotel-weather data using Auto Loader 
hotel_weather_stream_df = (
    spark.readStream
    .format("delta")
    .table("silver.hotel_weather_processed")
)

display(hotel_weather_stream_df)
```

```python
from AESUtils import AESService
import os

# Get AES key
try:
    aes_key = os.getenv("AES_KEY")
except:
    aes_key = "my-secure-encryption-key"

# Create AES service
aes_service = AESService(aes_key)
```

```python
PII_COLUMNS = ["name", "address"]
decrypted_df=aes_service.decrypt_pii_columns(hotel_weather_stream_df, PII_COLUMNS)
display(decrypted_df)
```

# Calculate the number of distinct hotels for each day in each city and country


#For each day (wthr_date) in each city in each country

```python
from pyspark.sql.functions import (
    col,approx_count_distinct,avg,max,min
)

gold_df = decrypted_df.groupBy(
    "country","city","wthr_date"
).agg(
    approx_count_distinct("id").alias("distinct_hotels"),
    
    avg("avg_tmpr_c").alias("average_temperature_c"),
    avg("avg_tmpr_f").alias("average_temperature_f"),
    
    max("avg_tmpr_c").alias("max_temperature_c"),
    min("avg_tmpr_c").alias("min_temperature_c"),
    
    max("avg_tmpr_f").alias("max_temperature_f"),
    min("avg_tmpr_f").alias("min_temperature_f")
).withColumn(
    "temperature_difference_c",
    col("max_temperature_c") - col("min_temperature_c")
).withColumn(
    "temperature_difference_f",
    col("max_temperature_f") - col("min_temperature_f")
)
```

```python
gold_df.printSchema()
```

```python
(gold_df.writeStream
    .format("delta")
    .option("checkpointLocation", CHECKPOINT_LOCATION)
    .outputMode("complete")
    .toTable("gold.hotel_weather_metrics")
)
```
