Reading data and configuring

```python
# Storage and paths
STORAGE_ACCOUNT = "stdevwesteuropepqa5"
CONTAINER_NAME = "data"
BASE_PATH = f"wasbs://{CONTAINER_NAME}@{STORAGE_ACCOUNT}.blob.core.windows.net"
HOTEL_WEATHER_SOURCE_PATH = f"{BASE_PATH}/hotel-weather"

# Auto Loader metadata locations (in DBFS)
SCHEMA_LOCATION = f"{BASE_PATH}/schema/bronze"
CHECKPOINT_LOCATION = f"{BASE_PATH}/checkpoint/bronze"

# Read hotel-weather data using Auto Loader 
hotel_weather_stream_df = (
    spark.readStream
    .format("cloudFiles")
    .option("cloudFiles.format", "parquet")  
    .option("cloudFiles.schemaLocation", SCHEMA_LOCATION)
    .load(HOTEL_WEATHER_SOURCE_PATH)
)

display(hotel_weather_stream_df)
```

Importing and Initialising AES service 

```python
from AESUtils import AESService
import os

# Get AES key
try:
    aes_key = os.getenv("AES_KEY")
except:
    aes_key = "my-secure-encryption-key"

# Create AES service
aes_service = AESService(aes_key)
```

Encrypt the data

```python
PII_COLUMNS = ["name", "address"]
encrypted_df = aes_service.encrypt_pii_columns(hotel_weather_stream_df, PII_COLUMNS)
display(encrypted_df)
```

write data to bronze notebook again

```python
(encrypted_df.writeStream
    .format("delta")
    .option("checkpointLocation", CHECKPOINT_LOCATION)
    .outputMode("append")
    .toTable("bronze.hotel_weather_raw")
)
```
