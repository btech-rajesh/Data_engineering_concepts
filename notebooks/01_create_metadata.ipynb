## Config Path and Schema 

```python
# Azure Storage Configuration
STORAGE_ACCOUNT = "tfdataaccount"
CONTAINER_NAME = "data"

# Base path
BASE_PATH = f"wasbs://{CONTAINER_NAME}@{STORAGE_ACCOUNT}.blob.core.windows.net"

# Checking access key
try:
    contents = dbutils.fs.ls(BASE_PATH)
    print(f"Connection successful!")
except Exception as e:
    print(f"Connection failed: {e}")

# Schema
BRONZE = "bronze"
SILVER = "silver"
GOLD = "gold"

# Schema Path
BRONZE_PATH = f"{BASE_PATH}/bronze"
SILVER_PATH = f"{BASE_PATH}/silver"
GOLD_PATH = f"{BASE_PATH}/gold"
```

## Creating schema

```python
def create_schema(schema_name, location, comment=None):
    schema_comment = f"COMMENT '{comment}'" if comment else ""
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS {schema_name} {schema_comment} LOCATION '{location}'")
    print(f"Schema {schema_name} created")

create_schema(BRONZE, BRONZE_PATH, "Bronze layer: Raw data")
create_schema(SILVER, SILVER_PATH, "Silver layer: Cleaned data")
create_schema(GOLD, GOLD_PATH, "Gold layer: Aggregated data")
```

## Print Schema

```python
# List of all schema
display(spark.sql("SHOW SCHEMAS"))

# Details of Schemas
for schema in [BRONZE, SILVER, GOLD]:
    print(f"---------{schema.upper()}----------")
    display(spark.sql(f"DESCRIBE schema {schema}"))
```
